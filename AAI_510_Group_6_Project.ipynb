```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Set plot style for better visualization
plt.style.use('seaborn-v0_8') # Use seaborn style for cleaner plots
```

```python
# Load the Rossmann Store Sales dataset
train_data = pd.read_csv('train.csv', low_memory=False)  # Load training data (sales data)
store_data = pd.read_csv('store.csv')  # Load store metadata

# Display basic information about the dataset
print("=== Train Data Overview ===")
print(train_data.info())  # Data types, missing values, and memory usage
print("\n=== Store Data Overview ===")
print(store_data.info())
```

```python
# Check for missing values
print("\n=== Missing Values in Train Data ===")
print(train_data.isnull().sum())
print("\n=== Missing Values in Store Data ===")
print(store_data.isnull().sum())
```

```python
# Basic statistics for numerical columns
print("\n=== Train Data Description ===")
print(train_data.describe())
```

```python
# Initial visualization: Sales trend over time
# Convert 'Date' column to datetime
train_data['Date'] = pd.to_datetime(train_data['Date'])
```

```python
# Plot sales over time for a sample store (e.g., Store 1)
store_1_data = train_data[train_data['Store'] == 1]
plt.figure(figsize=(10, 6))
plt.plot(store_1_data['Date'], store_1_data['Sales'], label='Daily Sales (Store 1)')
plt.title('Sales Trend for Store 1')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Initial visualization: Sales distribution
plt.figure(figsize=(8, 6))
sns.histplot(train_data['Sales'], bins=50, kde=True)
plt.title('Distribution of Sales Across All Stores')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```
